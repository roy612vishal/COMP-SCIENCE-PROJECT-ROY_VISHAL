```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
# from sklearn.svm import SVC # Uncomment and use this for Support Vector Machine
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, roc_auc_score, auc
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
import os

# Set display options
pd.set_option('display.max_columns', None)
warnings.filterwarnings("ignore") # Suppress warnings for cleaner output

# For Jupyter Notebooks (comment out if not in a Jupyter environment)
# You may need to uncomment the following lines if you are running this in a Jupyter Notebook
# try:
#     get_ipython().run_line_magic('matplotlib', 'inline')
# except NameError:
#     pass # Not in a Jupyter environment

# Configure Plotly (optional, for interactive plots if needed)
import plotly.io as pio
# pio.renderers.default = 'notebook' # Uncomment if in a notebook, adjust as needed for your environment


def load_datasets():
    """
    Loads the datasets from CSV and Excel files using raw strings for paths.
    Includes specific error handling for each file and diagnostic prints.
    Returns:
        gdsc_data, cell_line_data, compound_data: Loaded pandas DataFrames if successful, otherwise returns None for failed loads.
    """
    print("Attempting to load datasets...")
    gdsc_data = None
    cell_line_data = None
    compound_data = None

    # Load GDSC data
    try:
        # IMPORTANT: Use raw strings (r"...") for file paths to correctly handle backslashes
        # Double-check this path is correct for your system
        gdsc_data_path = r"C:\Users\visha\Downloads\PROJECT-COMPUTER-main\PROJECT-COMPUTER-main\GDSC2-dataset.csv\GDSC2-dataset.csv"
        gdsc_data = pd.read_csv(gdsc_data_path)
        print(f"Successfully loaded GDSC data from {gdsc_data_path}. Shape: {gdsc_data.shape}")
    except FileNotFoundError:
        print(f"Error: GDSC data file not found at {gdsc_data_path}. Please check the path.")
    except Exception as e:
        print(f"Error loading GDSC data from {gdsc_data_path}: {e}")

    # Load Cell Line data
    try:
        # IMPORTANT: Use raw strings (r"...") for file paths
        # Double-check this path and sheet name are correct for your system
        cell_line_data_path = r"C:\Users\visha\Downloads\PROJECT-COMPUTER-main\PROJECT-COMPUTER-main\Cell_Lines_Details.xlsx"
        cell_line_data = pd.read_excel(cell_line_data_path, sheet_name='Cell line details')
        print(f"Successfully loaded Cell Line data from {cell_line_data_path}. Sheet: 'Cell line details'. Shape: {cell_line_data.shape}")
    except FileNotFoundError:
        print(f"Error: Cell Line data file not found at {cell_line_data_path}. Please check the path and filename.")
    except Exception as e:
        print(f"Error loading Cell Line data from {cell_line_data_path}: {e}")

    # Load Compound data
    try:
        # IMPORTANT: Use raw strings (r"...") for file paths
        # Double-check this path is correct for your system
        compound_data_path = r"C:\Users\visha\Downloads\PROJECT-COMPUTER-main\PROJECT-COMPUTER-main\Compounds-annotation.csv"
        compound_data = pd.read_csv(compound_data_path)
        print(f"Successfully loaded Compound data from {compound_data_path}. Shape: {compound_data.shape}")
    except FileNotFoundError:
        print(f"Error: Compound data file not found at {compound_data_path}. Please check the path.")
    except Exception as e:
        print(f"Error loading Compound data from {compound_data_path}: {e}")

    # Check if all essential dataframes were loaded
    if gdsc_data is not None and cell_line_data is not None and compound_data is not None:
        print("\nAll essential datasets loaded successfully!")
    else:
        print("\nOne or more essential datasets failed to load. Merging and analysis may be affected.")


    return gdsc_data, cell_line_data, compound_data


def inspect_data(gdsc_data, cell_line_data, compound_data):
    """
    Inspects the shape and columns of the input DataFrames if they are not None.
    Args:
        gdsc_data, cell_line_data, compound_data: Input pandas DataFrames (or None).
    """
    print("\n=== Dataset Information ===")
    if gdsc_data is not None:
        print("GDSC Data Shape:", gdsc_data.shape)
        # print("GDSC Data Columns:", gdsc_data.columns.tolist()) # Uncomment if you want to see all columns
    else:
        print("GDSC Data not loaded.")

    if cell_line_data is not None:
        print("\nCell Line Data Shape:", cell_line_data.shape)
        # print("Cell Line Data Columns:", cell_line_data.columns.tolist()) # Uncomment if you want to see all columns
    else:
        print("\nCell Line Data not loaded.")

    if compound_data is not None:
        print("\nCompound Data Shape:", compound_data.shape)
        # print("Compound Data Columns:", compound_data.columns.tolist()) # Uncomment if you want to see all columns
    else:
        print("\nCompound Data not loaded.")


def clean_and_merge_data(gdsc_data, cell_line_data):
    """
    Cleans and merges the gdsc_data and cell_line_data DataFrames on common columns.
    Args:
        gdsc_data, cell_line_data: Input pandas DataFrames.
    Returns:
        merged_data: Merged pandas DataFrame or None if input is missing or merge fails.
    """
    if gdsc_data is None or cell_line_data is None:
        print("\nSkipping first merge (GDSC + Cell Line) due to missing input data.")
        return None

    print("\n=== Data Cleaning and Merging (GDSC + Cell Line) ===")
    # Rename columns for consistency
    if 'Sanger Cell Line Name' in cell_line_data.columns:
        cell_line_data.rename(columns={'Sanger Cell Line Name': 'CELL_LINE_NAME'}, inplace=True)
    if 'COSMIC identifier' in cell_line_data.columns:
        cell_line_data.rename(columns={'COSMIC identifier': 'COSMIC_ID'}, inplace=True)


    # Check for common columns for the first merge
    common_columns = list(set(gdsc_data.columns).intersection(set(cell_line_data.columns)))
    print("\nCommon columns for first merge (GDSC + Cell Line):", common_columns)

    if not common_columns: # Check if the list is empty
        print("\nCannot merge GDSC and Cell Line data - no common columns found")
        return None

    print("\nAttempting first merge...")
    try:
        # First merge (GDSC data with Cell Line details)
        merged_data = pd.merge(gdsc_data, cell_line_data, on=common_columns, how='left')
        print("First merge successful. Shape:", merged_data.shape)
        return merged_data
    except Exception as e:
        print(f"Error during first merge: {e}")
        return None # Return None on error


def merge_compound_data(merged_data, compound_data):
    """
    Merges the compound data into the merged dataset.
    Args:
        merged_data: DataFrame resulting from the first merge (GDSC + Cell Line).
        compound_data: compound data DataFrame.
    Returns:
        final_data: Complete merged DataFrame or None if input is missing or merge fails.
    """
    if merged_data is None or compound_data is None:
        print("\nSkipping second merge (with Compounds) due to missing input data from previous steps.")
        return None

    print("\n=== Data Merging (with Compounds) ===")
    # Check for common columns for the second merge
    common_compound_column = 'DRUG_ID' # Assuming 'DRUG_ID' is the common column with compound data
    if common_compound_column not in merged_data.columns or common_compound_column not in compound_data.columns:
         print(f"Error: Common column '{common_compound_column}' not found in both dataframes for the second merge.")
         # Attempt to find alternatives or raise error
         common_compound_columns = list(set(merged_data.columns).intersection(set(compound_data.columns)))
         if common_compound_columns:
             print(f"Found alternative common columns: {common_compound_columns}. Using the first one: {common_compound_columns[0]}")
             common_compound_column = common_compound_columns[0]
         else:
             print(f"No suitable common column found for second merge with compound data. Expected '{common_compound_column}'.")
             return None # Return None if merge column is definitively missing


    print(f"\nAttempting second merge on column '{common_compound_column}'...")
    try:
        # Second merge (with Compound annotation data)
        final_data = pd.merge(merged_data, compound_data, on=common_compound_column, how='left')
        print("Second merge successful. Final shape:", final_data.shape)

        # Basic data checks on merged data
        print("\nMissing values after merging:")
        print(final_data.isnull().sum().nlargest(10)) # Show top 10 columns with most missing values

        print("\nDuplicate rows after merging:", final_data.duplicated().sum())
        return final_data
    except Exception as e:
        print(f"Error during second merge: {e}")
        return None # Return None on error


def preprocess_genomic_data(final_data):
    """
    Preprocesses the merged genomic and drug sensitivity data for ML.
    Handles missing target values, categorical features, and missing numerical values.
    Args:
        final_data: Merged pandas DataFrame from previous steps.
    Returns:
        X: Feature DataFrame ready for ML.
        y: Target variable Series (encoded as numeric).
        feature_names: List of feature names.
        target_names: List of original target class names (e.g., ['Resistant', 'Sensitive']).
                 Returns None for any if preprocessing fails or data is insufficient.
    """
    if final_data is None or final_data.empty:
        print("\nSkipping preprocessing due to missing or empty input data.")
        return None, None, None, None

    print("\n=== Data Preprocessing ===")

    # Identify target variable (e.g., predicting sensitivity based on a threshold)
    # Let's create a binary target: 'Sensitive' or 'Resistant' based on LN_IC50
    target_column = 'LN_IC50' # Use LN_IC50 as the basis for the target

    if target_column not in final_data.columns:
        print(f"Target column '{target_column}' not found in the data. Cannot proceed with preprocessing.")
        return None, None, None, None

    # Drop rows where the target variable is missing
    initial_rows = final_data.shape[0]
    final_data_processed = final_data.dropna(subset=[target_column]).copy() # Create a copy to avoid modifying original
    rows_dropped = initial_rows - final_data_processed.shape[0]
    if rows_dropped > 0:
        print(f"Dropped {rows_dropped} rows with missing target variable ('{target_column}').")
    # else:
        # print(f"No missing values found in the target variable ('{target_column}').") # Keep output cleaner if no drops


    if final_data_processed.empty:
        print("DataFrame is empty after dropping rows with missing target. Cannot proceed with preprocessing.")
        return None, None, None, None

    # Define a target threshold. Using the median as a simple example:
    target_threshold = final_data_processed[target_column].median()
    print(f"Using median of '{target_column}' ({target_threshold:.4f}) as the target threshold.")


    # Create a binary target variable based on the threshold
    final_data_processed['DRUG_RESPONSE'] = final_data_processed[target_column].apply(
        lambda x: 'Sensitive' if x <= target_threshold else 'Resistant'
    )
    y = final_data_processed['DRUG_RESPONSE']
    print(f"Created binary target variable 'DRUG_RESPONSE' based on '{target_column}' with threshold {target_threshold:.4f}.")
    print("Target distribution after creating binary target:\n", y.value_counts())

    # Check if target has more than one class after creating binary target
    if len(y.unique()) < 2:
        print(f"Target variable 'DRUG_RESPONSE' has only one unique class after applying threshold ({y.unique()[0]}). Cannot perform classification.")
        print("Consider adjusting the target threshold.")
        return None, None, None, None


    # Select features (Exclude ID columns, original target variables, and the new target variable)
    feature_columns = [col for col in final_data_processed.columns if col not in ['COSMIC_ID', 'CELL_LINE_NAME', 'DRUG_ID', 'DRUG_NAME',
                                                                        'LN_IC50', 'AUC', 'Z_SCORE', 'DRUG_RESPONSE']]
    X = final_data_processed[feature_columns].copy()
    print(f"\nSelected {X.shape[1]} features for modeling.")
    # print("Selected features:", X.columns.tolist()) # Too long to print all feature names


    # Separate categorical and numerical features
    categorical_cols = X.select_dtypes(include=['object', 'category']).columns
    numerical_cols = X.select_dtypes(include=np.number).columns

    print(f"Identified {len(categorical_cols)} categorical features and {len(numerical_cols)} numerical features.")

    # Handle Missing Values
    # Using SimpleImputer for demonstration; KNNImputer could be used for numerical features
    if numerical_cols.size > 0:
        print("Handling missing values in numerical features using SimpleImputer (mean)...")
        numerical_imputer = SimpleImputer(strategy='mean')
        X[numerical_cols] = numerical_imputer.fit_transform(X[numerical_cols])
        print(f"Missing values in numerical features after imputation: {X[numerical_cols].isnull().sum().sum()}")

    if categorical_cols.size > 0:
        print("Handling missing values in categorical features using SimpleImputer (constant 'Missing')...")
        categorical_imputer = SimpleImputer(strategy='constant', fill_value='Missing')
        # Ensure columns are object type before imputation if they aren't already
        for col in categorical_cols:
             X[col] = X[col].astype('object')
        X[categorical_cols] = categorical_imputer.fit_transform(X[categorical_cols])
        print(f"Missing values in categorical features after imputation: {X[categorical_cols].isnull().sum().sum()}")


    # Handle Categorical Features (One-Hot Encoding)
    if categorical_cols.size > 0:
        print("Handling categorical features using One-Hot Encoding...")
        # Store column names before one-hot encoding
        original_cols_before_dummies = X.columns.tolist()
        X = pd.get_dummies(X, columns=categorical_cols, dummy_na=False)
        # Identify the new numerical columns created from one-hot encoding
        newly_created_cols = [col for col in X.columns if col not in original_cols_before_dummies]
        print(f"Created {len(newly_created_cols)} new columns after One-Hot Encoding.")


    # Scale Numerical Features (StandardScaler)
    # Identify numerical columns *after* one-hot encoding for scaling
    # It's often better to scale *original* numerical columns.
    # Let's re-identify the original numerical columns that are still present in X
    original_numerical_cols_present = [col for col in numerical_cols if col in X.columns] # These were numerical *before* get_dummies

    if original_numerical_cols_present:
         print("Scaling original numerical features using StandardScaler...")
         scaler = StandardScaler()
         X[original_numerical_cols_present] = scaler.fit_transform(X[original_numerical_cols_present])
         print("Original numerical features scaled.")
    else:
         print("No original numerical columns found in the DataFrame after preprocessing. Skipping scaling.")

    # Get the final list of feature names
    final_feature_names = X.columns.tolist()

    # Check if X is empty after preprocessing
    if X.empty:
        print("\nFeature DataFrame is empty after preprocessing steps. Cannot proceed with model training.")
        return None, None, None, None


    # Encode the target variable as numeric (0 and 1) for ML models
    # Ensure y is a pandas Series for LabelEncoder
    if isinstance(y, pd.Series) and not pd.api.types.is_numeric_dtype(y):
        print("Encoding target variable using LabelEncoder...")
        label_encoder = LabelEncoder()
        y_encoded = label_encoder.fit_transform(y)
        target_class_names = list(label_encoder.classes_) # Get the class names corresponding to 0 and 1
        print(f"Target variable encoded. Classes: {target_class_names} -> {label_encoder.transform(target_class_names)}")
    else:
        y_encoded = y
        # Assuming if y is already numeric, the classes are 0 and 1 and we need to derive names if possible
        # This part might need adjustment based on how your numeric target is structured if not 0/1
        # If y is already numeric, let's try to get unique values to use as target names
        if isinstance(y, (pd.Series, np.ndarray)):
             target_class_names = sorted(np.unique(y).tolist())
             print(f"Target variable was already numeric. Unique values: {target_class_names}")
        else: # Fallback if y is not a Series or array (unexpected)
             target_class_names = ['Class 0', 'Class 1'] # Default names if already numeric
             print("Warning: Target variable format unexpected. Using default target names.")


    print("\nPreprocessing complete.")
    print("Feature matrix shape:", X.shape)
    print("Target variable shape:", y_encoded.shape)
    return X, y_encoded, final_feature_names, target_class_names


def train_and_evaluate_ml_model(X, y, feature_names, target_names):
    """
    Trains a basic Machine Learning classification model (Logistic Regression)
    and evaluates its performance with visualizations.
    Args:
        X: Feature DataFrame.
        y: Target variable Series (encoded as numeric).
        feature_names: List of names corresponding to the features in X.
        target_names: List of original target class names (e.g., ['Resistant', 'Sensitive']).
                 Expected to be a list of strings corresponding to encoded labels 0, 1...
    """
    # --- Input Data Validation ---
    # These checks are crucial to ensure we have data before proceeding
    if X is None or y is None or feature_names is None or target_names is None:
        print("\nInput data for model training is incomplete (None value received). Skipping training and evaluation.")
        return

    # Check if the dataframes/arrays are empty using .shape[0] (number of rows)
    # X and y should be pandas DataFrames/Series or NumPy arrays here
    if not hasattr(X, 'shape') or X.shape[0] == 0 or not hasattr(y, 'shape') or y.shape[0] == 0:
         print("\nInput data for model training is empty or invalid (zero rows/missing shape). Skipping training and evaluation.")
         return


    # Check if target variable has more than one class
    # This check is now also performed in preprocessing, but good to have here too
    if len(np.unique(y)) < 2:
        print(f"\nTarget variable has only one class ({np.unique(y)[0]}). Cannot perform classification. Skipping training and evaluation.")
        print("Consider adjusting the target threshold in preprocess_genomic_data to create a binary target.")
        return

    # Ensure target_names is a list and has the correct number of elements
    if not isinstance(target_names, list) or len(target_names) != len(np.unique(y)):
        print("\nWarning: target_names list does not match the number of unique classes in the target variable.")
        print("Using unique values from y as target names for evaluation reports.")
        # Attempt to regenerate target_names from unique values in y if possible
        try:
             target_names_for_report = sorted([str(name) for name in np.unique(y).tolist()])
             print(f"Using generated target names for report: {target_names_for_report}")
        except Exception:
             target_names_for_report = ['Class 0', 'Class 1'] # Fallback default
             print("Could not generate target names from unique y values. Using default.")
    else:
         target_names_for_report = [str(name) for name in target_names] # Ensure they are strings


    print("\n=== Model Development, Training, and Evaluation ===")

    # Split data into training and testing sets
    # Using stratified split to maintain target distribution in train and test sets
    print("Splitting data into training and testing sets...")
    try:
        # Ensure y is in a format acceptable by train_test_split stratify (Series or array)
        if isinstance(y, pd.Series):
            y_stratify = y
        elif isinstance(y, np.ndarray):
            y_stratify = pd.Series(y) # Convert to Series for value_counts and consistent stratify behavior
        else:
             print("Warning: Target variable format unexpected for stratification. Proceeding without stratification.")
             X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
             y_stratify = None # Disable stratified check prints later


        if y_stratify is not None:
             # Check if stratification is possible (requires at least 2 samples per class in each split)
             # This check is implicitly done by train_test_split but an explicit check can be helpful
             stratify_possible = True
             for class_val in y_stratify.unique():
                  if y_stratify.value_counts()[class_val] < 2: # Needs at least 2 for both train/test split (0.75*count >=1 and 0.25*count >=1 implies count >= 4, but minimum is 2 for split)
                       print(f"Warning: Class {class_val} has too few samples ({y_stratify.value_counts()[class_val]}) for stratification. train_test_split might fail or not stratify.")
                       # Decide whether to proceed without stratification or raise error
                       # For now, let train_test_split handle it and catch ValueError
                       # stratify_possible = False # Can set this flag and use non-stratified split


             X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y_stratify)
             print(f"Training target distribution:\n{pd.Series(y_train).value_counts()}")
             print(f"Testing target distribution:\n{pd.Series(y_test).value_counts()}")


    except ValueError as e:
        print(f"Error splitting data with stratification: {e}. This might happen if a class has too few samples for stratification.")
        print("Attempting to split data WITHOUT stratification...")
        try:
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
            print("Data split successfully WITHOUT stratification.")
            print(f"Training target distribution:\n{pd.Series(y_train).value_counts()}")
            print(f"Testing target distribution:\n{pd.Series(y_test).value_counts()}")
        except Exception as split_e:
             print(f"Error splitting data even without stratification: {split_e}")
             print("Skipping model training and evaluation.")
             return # Exit if splitting fails


    except Exception as e:
        print(f"An unexpected error occurred during data splitting: {e}")
        print("Skipping model training and evaluation.")
        return # Exit if splitting fails


    # Initialize and train a Logistic Regression model
    # You can replace LogisticRegression with SVC(probability=True, random_state=42) for SVM
    print("\nTraining Logistic Regression model...")
    try:
        model = LogisticRegression(max_iter=1000, random_state=42) # Increased max_iter for convergence
        model.fit(X_train, y_train)
        print("Model training complete.")
    except Exception as e:
        print(f"Error during model training: {e}")
        print("Model training failed. Please check the processed data (X, y) for issues like non-numeric values or infinite values.")
        return # Exit function if training fails


    # --- Model Evaluation ---
    print("\n=== Model Evaluation ===")
    # Ensure model is trained before predicting
    if 'model' not in locals() or model is None:
        print("Model was not trained successfully. Skipping evaluation.")
        return

    # Check if X_test is empty
    if X_test.shape[0] == 0:
        print("Test set is empty. Cannot perform evaluation.")
        return


    y_pred = model.predict(X_test)

    # Get probability estimates for the positive class (required for ROC curve)
    # If using SVC, ensure probability=True is set during initialization
    y_prob = None # Initialize y_prob to None
    has_predict_proba = False
    try:
        # Check if the model has predict_proba and if there are exactly two classes
        if hasattr(model, 'predict_proba') and len(model.classes_) == 2:
             # Find the index of the positive class (assuming 'Sensitive' is positive, or class with label 1)
             # Need to be careful about which class predict_proba[:, 1] refers to
             # It refers to the class with label 1 after LabelEncoding
             # Let's find which original class name corresponds to label 1
             try:
                 # Find the original name corresponding to encoded label 1
                 positive_class_name = target_names_for_report[1] # Assuming 1 is the positive label
                 print(f"Assuming positive class for ROC is '{positive_class_name}' (encoded label 1).")
             except Exception:
                 positive_class_name = 'Encoded Label 1'
                 print(f"Could not determine positive class name for ROC. Assuming 'Encoded Label 1'.")


             y_prob = model.predict_proba(X_test)[:, 1]
             has_predict_proba = True
             # print("Model supports predict_proba.") # Already printed


        elif hasattr(model, 'predict_proba') and len(model.classes_) != 2:
             print(f"Model has {len(model.classes_)} classes, predict_proba[:, 1] is suitable for binary classification. Skipping probability predictions.")
        else:
             print("Model does not support predict_proba. Cannot generate ROC curve.")

    except Exception as e:
        print(f"Error getting prediction probabilities: {e}")
        print("Skipping probability-based evaluation (e.g., ROC AUC).")
        y_prob = None
        has_predict_proba = False


    # Classification metrics
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nAccuracy: {accuracy:.4f}")
    print("\nClassification Report:")
    # classification_report needs target_names to display labels clearly
    # Ensure target_names_for_report corresponds to the encoded labels (0 and 1)
    try:
        print(classification_report(y_test, y_pred, target_names=target_names_for_report))
    except Exception as e:
        print(f"Error generating classification report: {e}")
        print("Printing classification report without target names:")
        # Fallback if target_names_for_report somehow caused an issue
        print(classification_report(y_test, y_pred))


    # Confusion Matrix Visualization
    print("\nGenerating Confusion Matrix...")
    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(8, 6))
    # Use target_names_for_report for the confusion matrix labels
    try:
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_names_for_report, yticklabels=target_names_for_report)
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.title('Confusion Matrix')
        plt.show()
    except Exception as e:
         print(f"Error generating confusion matrix plot: {e}")
         print("Skipping confusion matrix plot.")


    # ROC Curve and AUC
    if has_predict_proba and y_prob is not None:
        print("\nGenerating ROC Curve...")
        # roc_curve function needs the true binary labels and the probability estimates for the positive class
        # Ensure y_test is binary (0 or 1) and corresponds to y_prob
        # Check if y_test contains exactly two classes before plotting ROC
        if len(np.unique(y_test)) == 2:
            try:
                 fpr, tpr, thresholds = roc_curve(y_test, y_prob)
                 roc_auc = auc(fpr, tpr)

                 plt.figure(figsize=(8, 6))
                 plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
                 plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
                 plt.xlim([0.0, 1.0])
                 plt.ylim([0.0, 1.05])
                 plt.xlabel('False Positive Rate')
                 plt.ylabel('True Positive Rate')
                 plt.title('Receiver Operating Characteristic (ROC) Curve')
                 plt.legend(loc="lower right")
                 plt.show()

                 print(f"ROC AUC Score: {roc_auc:.4f}")
            except Exception as e:
                print(f"Error generating ROC curve plot: {e}")
                print("Skipping ROC curve plot.")

        else:
            print("ROC curve requires binary target variable (exactly 2 classes) in the test set. Skipping plot.")


    else:
        print("\nSkipping ROC Curve generation (model does not support probabilities or data issues).")


    # You can add other relevant visualizations here, e.g., feature importance if using models that support it
    # For Logistic Regression, coefficients can indicate feature importance, but need careful interpretation
    # if features were scaled and encoded.


# --- Main Execution Flow ---
if __name__ == "__main__":
    # 1. Data Acquisition and Integration
    # Added more specific error handling and diagnostic prints in load_datasets
    # Removed try/except with exit() here to allow granular error messages from load_datasets to show
    gdsc_data, cell_line_data, compound_data = load_datasets()

    # Check if essential data was loaded before proceeding
    # The individual load functions print errors, so we just check if they returned None
    if gdsc_data is None or cell_line_data is None or compound_data is None:
         print("\nEssential data not loaded. Cannot proceed with merging and analysis.")
         # Set subsequent data variables to None to ensure functions handle missing input
         merged_data = None
         final_data = None
         X_processed = None
         y_target_encoded = None
         feature_names = None
         target_class_names = None

    else:
        # 2. Data Inspection (only if data was loaded)
        inspect_data(gdsc_data, cell_line_data, compound_data)

        # 3. Data Cleaning and Merging
        merged_data = clean_and_merge_data(gdsc_data, cell_line_data)
        print(f"Value of merged_data after clean_and_merge_data: {type(merged_data)}")

        # Check if first merge was successful
        if merged_data is None:
             print("\nFirst merge failed. Cannot proceed with subsequent merges and analysis.")
             final_data = None # Set to None to ensure subsequent checks work
             X_processed = None
             y_target_encoded = None
             feature_names = None
             target_class_names = None
        else:
             final_data = merge_compound_data(merged_data, compound_data)
             print(f"Value of final_data after merge_compound_data: {type(final_data)}")

             # Check if final_data is valid before proceeding to preprocessing
             if final_data is None or final_data.empty:
                  print("\nFinal data is missing or empty after merging. Skipping preprocessing and model training.")
                  X_processed = None # Explicitly set to None so subsequent checks work
                  y_target_encoded = None
                  feature_names = None
                  target_class_names = None
             else:
                  # 4. Data Preprocessing (only if final_data is valid)
                  # Pass final_data to the preprocessing function
                  # The preprocess function returns processed data or None if issues occur
                  X_processed, y_target_encoded, feature_names, target_class_names = preprocess_genomic_data(final_data.copy()) # Use a copy to avoid modifying original


    # 5. Model Development, Training, and Evaluation with Visualizations
    # Pass processed features and target to the training function
    # train_and_evaluate_ml_model includes checks for None/empty data
    # This function will now be called even if preprocessing failed, but will skip training internally
    train_and_evaluate_ml_model(X_processed, y_target_encoded, feature_names, target_class_names)


    # 6. Exploratory Data Visualization (Optional - can be done on raw or processed data)
    # You can call safe_plotting(final_data) here if you still want to see the initial data distributions etc.
    # Note: safe_plotting also needs to handle None or empty input data gracefully if it's called when data failed to load/merge
    # safe_plotting(final_data.copy() if final_data is not None and not final_data.empty else pd.DataFrame()) # Example defensive call


    # 7. Save Processed Data (Optional - save final_data or X_processed/y_target if needed)
    # save_processed_data(final_data.copy() if final_data is not None and not final_data.empty else pd.DataFrame()) # Example defensive call

    print("\n=== Full Pipeline Execution Attempt Complete ===")
```
